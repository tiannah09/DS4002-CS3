```python
import pandas as pd
df = pd.read_csv("../data/initial_data.csv") # upload dataset
df.head()
```

```python
# drop unneccessary columns 
df = df.drop(columns=["state_event", "Datetime Event Began", "Datetime Restoration", "fips", 
                      "min_customers", "max_customers", "mean_customers"])
```

```python
df["hour"] = pd.to_datetime(df["start_time"]).dt.hour # convert start_time to a datetime object
```

```python
df.to_csv("../data/cleaned_data.csv", index=False) # save to new csv
```
